```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data= pd.read_csv("/content/StudentsPerformance.csv")
data.head()
```

```python
data.columns
```

```python
data.isnull()
```

```python
data.describe()
```

```python
gender_counts = data['gender'].value_counts()

plt.figure(figsize=(8, 6))
plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%', startangle=140, colors=['lightpink','skyblue'])
plt.title('Distribution of Students by Gender')
plt.show()
```

# Visualization 1: Distribution of Scores

```python
sns.set_style("whitegrid")

plt.figure(figsize=(18, 5))

# Math score distribution
plt.subplot(1, 3, 1)
sns.histplot(data['math score'], bins=20, kde=True)
plt.title('Distribution of Math Scores')

# Reading score distribution
plt.subplot(1, 3, 2)
sns.histplot(data['reading score'], bins=20, kde=True)
plt.title('Distribution of Reading Scores')

# Writing score distribution
plt.subplot(1, 3, 3)
sns.histplot(data=['writing score'], bins=20, kde=True)
plt.title('Distribution of Writing Scores')

plt.tight_layout()
plt.show()
```

# Visualization 2: Gender Performance Comparison

```python
plt.figure(figsize=(10, 6))
sns.barplot(data=data, x='gender', y='math score')
plt.title('Average Math Scores by Gender')
plt.ylabel('Average Score')
plt.show()
```

# Visualization 3: Parental Education Level Impact

```python
plt.figure(figsize=(14, 7))
sns.barplot(data=data, x='parental level of education', y='math score', errorbar=None)
plt.title('Average Math Scores by Parental Education Level')
plt.xticks(rotation=45)
plt.ylabel('Average Score')
plt.show()
```

# Visualization 4: Lunch Program Comparison


```python
plt.figure(figsize=(10, 6))
sns.barplot(data=data, x='lunch', y='math score', errorbar=None)
plt.title('Average Math Scores by Lunch Program')
plt.ylabel('Average Score')
plt.show()
```

# Visualization 5: Test Preparation Course Effectiveness


```python
plt.figure(figsize=(10, 6))
sns.barplot(data=data, x='test preparation course', y='math score', errorbar=None)
plt.title('Average Math Scores by Test Preparation Course Status')
plt.ylabel('Average Score')
plt.show()
```

```python
# Visualization: Score Correlations
# Using pairplot to visualize the relationships between math, reading, and writing scores

sns.pairplot(data[['math score', 'reading score', 'writing score']], kind='reg', plot_kws={'line_kws':{'color':'red'}})
plt.suptitle('Score Correlations Among Math, Reading, and Writing', y=1.02)
plt.show()
```

```python
# Visualization: Heatmap of Score Correlations
corr_matrix = data[['math score', 'reading score', 'writing score']].corr()

plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix of Scores')
plt.show()

```
